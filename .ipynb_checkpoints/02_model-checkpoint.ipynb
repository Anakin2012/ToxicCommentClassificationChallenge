```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import nltk
from nltk import word_tokenize
from sklearn import model_selection
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = pd.read_csv('csv/train_01.csv')
#test = pd.read_csv('csv/test_01.csv')
```

```python
data = data.dropna(how='any',axis=0) 
#test = test.dropna(how='any', axis=0)
```

```python
data_text = data['comment_text'].values.astype('U')
#test_text = test['comment_text'].values.astype('U')
```

```python
y = data.drop(labels = ['id', 'comment_text'], axis=1)
```

```python
y.sum()
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(data, y, test_size = 0.33, random_state=7)
```

```python
vectorizer = TfidfVectorizer(strip_accents='unicode', tokenizer=word_tokenize, analyzer='word', ngram_range=(1,3), norm='l2', max_features = 20000)
```

```python
vectorizer.fit(X_train['comment_text'])
```

```python
X_train = vectorizer.transform(X_train['comment_text'])
X_test = vectorizer.transform(X_test['comment_text'])
```

```python
vectorizer.get_feature_names()
```

```python
vectorizer.vocabulary_
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC

clf = OneVsRestClassifier(LinearSVC(), n_jobs=1)
```

```python
clf.fit(X_train, y_train)
```

```python
score = clf.score(X_train, y_train)
```
